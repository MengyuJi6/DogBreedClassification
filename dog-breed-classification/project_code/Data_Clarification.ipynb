# Data Clarification

```
import numpy as np
import pandas as pd
from PIL import Image
```

```
df_train = pd.read_csv('labels.csv', delimiter=',', low_memory=False) 
df_train.head()
```

## Breed Metrics

See how many unique breeds do we have in this dataset:

```
unique_breeds = df_train.breed.unique()
len(unique_breeds)
```

Show the most frequent breeds and the least frequent breeds:

```
gr_labels = df_train.groupby("breed").count()
gr_labels = gr_labels.rename(columns = {"id" : "count"})
gr_labels = gr_labels.sort_values("count", ascending=False)
```

```
gr_labels.head()
```

```
gr_labels.tail()
```

## Training Data Visualization

```
ax=pd.value_counts(df_train['breed'],ascending=True).plot(kind='barh',
                                                       fontsize="10",
                                                       title="Class Distribution",
                                                       figsize=(20,40))
ax.set(xlabel="Images per class", ylabel="Classes")
ax.xaxis.label.set_size(15)
ax.yaxis.label.set_size(15)
ax.title.set_size(20)
```

```
ax=pd.value_counts(df_train['breed'],ascending=True).plot(kind='pie',
                                                       fontsize="10",
                                                       title="Class Distribution",
                                                       figsize=(20,20))
ax.set(xlabel="Images per class", ylabel="Classes")
ax.xaxis.label.set_size(15)
ax.yaxis.label.set_size(15)
ax.title.set_size(20)
```
